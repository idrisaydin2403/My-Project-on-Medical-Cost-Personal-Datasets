```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:\\Users\\İdris Aydın\\Desktop\\insurance.csv")
df
```

```python
number_of_male = len(df[df['sex'] == 'male'])
number_of_female = len(df[df['sex'] == 'female'])

ratio_male = (number_of_male / len(df)) * 100
ratio_female = (number_of_female / len(df)) * 100

print("\nNUMBER OF MALE PARTICIPANTS:", number_of_male)
print("NUMBER OF FEMALE PARTICIPANTS:", number_of_female,"\n")

plt.bar(['Male\n%'+str(round(ratio_male,2)),'Female\n%' + str(round(ratio_female,2))], [ratio_male, ratio_female], width = 0.2, color = ['blue','hotpink'])
plt.title("PERCENTAGE OF PARTICIPANTS\n")
plt.xlabel("GENDER")
plt.ylabel("PERCENTAGE\n")

plt.show()
```

```python
number_of_young = len(df[(df['age'] >= 18) & (df['age'] < 30)])
number_of_adult = len(df[(df['age'] >= 30) & (df['age'] < 60)])
number_of_old = len(df[df['age'] >= 60])

print("\nNUMBER OF YOUNG:", number_of_young)
print("NUMBER OF ADULT:", number_of_adult)
print("NUMBER OF OLD:", number_of_old,"\n\n")

ratio_young = (number_of_young / len(df)) * 100
ratio_adult = (number_of_adult / len(df)) * 100
ratio_old = (number_of_old / len(df)) * 100

plt.bar(['Young\n%' + str(round(ratio_young, 2)) + "\n\n[18-30)", 'Adult\n%' + str(round(ratio_adult, 2)) + "\n\n[30-60)",'Old\n%' + str(round(ratio_old, 2)) + "\n\n(60+]"], [ratio_young, ratio_adult, ratio_old], width = 0.4, color = 'gray')
plt.title("PERCENTAGE OF AGE CATEGORIES\n")
plt.xlabel("\nAGE CATEGORIES")
plt.ylabel("PERCENTAGE\n")

plt.show()
```

```python
average_male_bmi = np.mean(df[df['sex'] == 'male']['bmi'])
average_female_bmi = np.mean(df[df['sex'] == 'female']['bmi'])

print("\nAVERAGE BMI VALUE FOR MALE:", round(average_male_bmi, 2))
print("AVERAGE BMI VALUE FOR FEMALE:", round(average_female_bmi, 2),"\n\n")

plt.bar(['Male\n' + str(round(average_male_bmi, 2)),'Female\n' + str(round(average_female_bmi, 2))], [average_male_bmi, average_female_bmi], width = 0.2, color = ['blue','red'])
plt.title("AVERAGE BMI VALUES BY GENDER CATEGORY\n")
plt.xlabel("\nGENDER")
plt.ylabel("BMI VALUES\n")

plt.show()
```

```python
young_bmi = np.mean(df[((df['age'] >= 18) & (df['age'] < 30))]['bmi'])
adult_bmi = np.mean(df[((df['age'] >= 30) & (df['age'] < 60))]['bmi'])
old_bmi = np.mean(df[df['age'] >= 60]['bmi'])

print("\nAVERAGE BMI VALUE FOR YOUNG:", round(young_bmi,2))
print("AVERAGE BMI VALUE FOR ADULT:", round(adult_bmi, 2))
print("AVERAGE BMI VALUE FOR OLD:", round(old_bmi, 2) ,"\n\n")

colors = plt.cm.Blues(np.linspace(0.3, 0.5, 3))

plt.bar(['Young\n' + str(round(young_bmi, 2)),'Adult\n' + str(round(adult_bmi, 2)),'Old\n' + str(round(old_bmi,2))], [young_bmi, adult_bmi, old_bmi], width = 0.3, color = colors)
plt.title("AVERAGE BMI VALUE BY AGE CATEGORY\n")
plt.xlabel("\nAGE CATEGORY")
plt.ylabel("BMI VALUES\n")

plt.show()
```

```python
smokerNumber = len(df[df['smoker'] == 'yes'])
young_smoker = len(df[((df['age'] >= 18) & (df['age'] < 30)) & (df['smoker'] == 'yes')])
adult_smoker = len(df[((df['age'] >= 30) & (df['age'] < 60)) & (df['smoker'] == 'yes')])
old_smoker = len(df[(df['age'] > 60) & (df['smoker'] == 'yes')])

print("\nNUMBER OF YOUNG SMOKER:", young_smoker)
print("NUMBER OF ADULT SMOKER:", adult_smoker)
print("NUMBER OF OLD SMOKER:", old_smoker,"\n\n")

ratio_young_smoker = (young_smoker / smokerNumber) * 100
ratio_adult_smoker = (adult_smoker / smokerNumber) * 100
ratio_old_smoker = (old_smoker / smokerNumber) * 100

plt.bar(['Young\n%' + str(round(ratio_young_smoker,2)) ,'Adult\n%' + str(round(ratio_adult_smoker, 2)) ,'Old\n%' + str(round(ratio_old_smoker, 2))], [ratio_young_smoker, ratio_adult_smoker, ratio_old_smoker], width = 0.3, color = plt.cm.Reds(np.array([ratio_young_smoker, ratio_adult_smoker, ratio_old_smoker]) / 75))
plt.title("PERCENTAGE OF SMOKERS BY AGE CATEGORY\n")
plt.xlabel("\nAGE CATEGORY")
plt.ylabel("PERCENTAGE\n")

# We see a slight increase in BMI value as age increases.

plt.show()
```

```python
young_male_smoker = len(df[((df['age'] >= 18) & (df['age'] < 30)) & (df['smoker'] == 'yes') & (df['sex'] == 'male')])
young_female_smoker = len(df[((df['age'] >= 18) & (df['age'] < 30)) & (df['smoker'] == 'yes') & (df['sex'] == 'female')])

adult_male_smoker = len(df[((df['age'] >= 30) & (df['age'] < 60)) & (df['smoker'] == 'yes') & (df['sex'] == 'male')])
adult_female_smoker = len(df[((df['age'] >= 30) & (df['age'] < 60)) & (df['smoker'] == 'yes') & (df['sex'] == 'female')])

old_male_smoker = len(df[(df['age'] >= 60) & (df['smoker'] == 'yes') & (df['sex'] == 'male')])
old_female_smoker = len(df[(df['age'] >= 60) & (df['smoker'] == 'yes') & (df['sex'] == 'female')])

age_categories = ['Young','Adult','Old']
sub_categories = ['Male','Female']
percentages = [[(young_male_smoker / (young_male_smoker + young_female_smoker)) * 100, (young_female_smoker / (young_male_smoker + young_female_smoker)) * 100], [(adult_male_smoker / (adult_male_smoker + adult_female_smoker)) * 100, (adult_female_smoker / (adult_male_smoker + adult_female_smoker)) * 100], [(old_male_smoker / (old_male_smoker + old_female_smoker)) * 100, (old_female_smoker / (old_female_smoker + old_male_smoker)) * 100]]

bar_width = 0.25
index = np.arange(len(age_categories))

colors_male = plt.cm.Blues(0.7)
colors_female = plt.cm.Reds(0.4)

fig, ax = plt.subplots()
bar1 = ax.bar(index, [p[0] for p in percentages], bar_width, label = 'Male', color = colors_male)
bar2 = ax.bar(index + bar_width, [p[1] for p in percentages], bar_width, label = 'Female', color = colors_female)

ax.set_xlabel('\nAGE GROUPS')
ax.set_ylabel('PERCENTAGE\n')
ax.set_title('SMOKING DISTRIBUTION BY AGE AND GENDER\n')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(age_categories)
ax.legend()

# We observe that smoking occurs mostly in adults.

plt.show()
```
